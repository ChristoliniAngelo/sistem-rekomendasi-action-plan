```python
import pandas as pd
```

```python
df = pd.read_excel('data_fix.xlsx', header = 0)

df
```

```python
keywords_i = input('Masukkan 1-5 kata sebagai keyword: ')

keywords_i = keywords_i.lower()

list_ki = keywords_i.split(' ')

def check_keywords(list_ki, title):
    for word in list_ki:
        if word not in title.lower():
            return False
    return True

i = 0
for index, row in df.iterrows():
    # Check if all keywords exist, then provide improvement recommendations
    if check_keywords(list_ki, row['Project Title']):
        print(row['Improve'])
        print('-------------------------------------------------------------------------')
        i += 1
    # Limit 3 results being displayed
    if i >= 3:
        break
    
```
